```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from routines import ReadCOSMOthermIsotherm


inp_str = 'isotherms/raw/'
param_str = 'TZVP'
fgases = ['difluoromethane', '1,1,1,2-tetrafluoroethane', 'pentafluoroethane']
fgaslabels = ['R32', 'R134a', 'R125']

T = 303.15
```

# [TBA][NFS]:PFPA (1:2)

```python
des_str = '[TBA][NFS]:PFPA (1:2)'
des_inp = 'tba_nfs_pfpa'
rdes_str = '1_2'

plt.figure(figsize=(20, 8), dpi=100)
plt.suptitle('Solubility Curves in '+des_str)
for i in range(len(fgases)):
    COSMO_data = ReadCOSMOthermIsotherm(inp_str+des_inp+'/'+rdes_str+'__SolubilityIsotherms-'+param_str+'.tab')
    fgas_ndx = i
    fgas, fgaslabel = fgases[fgas_ndx], fgaslabels[fgas_ndx]
    X1, Y1, P = np.array(COSMO_data[(T, fgas)][0]), np.array(COSMO_data[(T, fgas)][1]), np.array(COSMO_data[(T, fgas)][2])
    plt.subplot(2, 4, i+1)
    plt.plot(X1, P, '-', label='%.2f K' % T, linewidth=2), plt.plot(Y1, P, '-', linewidth=2)
    plt.xlabel(r'$x_1$ / mol$\cdot$mol$^{-1}$'), plt.ylabel(r'$P$ / MPa')
    plt.text(0.05, 4.7, fgaslabels[i], fontsize=12, ha='left')
    plt.xlim(0, 1), plt.ylim(0, 5)
    np.savetxt('isotherms/'+fgaslabel+'+'+des_inp+'['+rdes_str+'].txt', np.array([X1, Y1, P]).T, header='# x1\t y1\t P', delimiter='\t')
plt.show()
```
